```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('Wine.csv')
X = df.drop(columns=['Customer_Segment'])
y = df['Customer_Segment']
```

```python
plt.figure(figsize=(8, 5))
for segment in y.unique():
    plt.scatter(X[y == segment]['Alcohol'], X[y == segment]['Malic_Acid'], label=f'Segment {segment}', alpha=0.7)
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.legend()
plt.title('Original Data')
plt.show()
```

```python
# Standardize features and apply PCA

X_scaled = StandardScaler().fit_transform(X)
X_pca = PCA(n_components=2).fit_transform(X_scaled)
```

```python
# Scatter plot after PCA
plt.figure(figsize=(8, 5))
for segment in y.unique():
    plt.scatter(X_pca[y == segment, 0], X_pca[y == segment, 1], label=f'Segment {segment}', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA: Wine Dataset')
plt.show()
```
